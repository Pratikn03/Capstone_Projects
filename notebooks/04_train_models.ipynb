This notebook trains baselines + GBM (Week 1) and the full GBM/LSTM/TCN suite (Week 2+).

```
import subprocess

# Baselines + GBM per target
for target in ['load_mw', 'wind_mw', 'solar_mw']:
    subprocess.run([
        'python', '-m', 'gridpulse.forecasting.train_baseline',
        '--features', 'data/processed/features.parquet',
        '--splits', 'data/processed/splits',
        '--target', target
    ], check=True)
```

```
import subprocess

# Full training (GBM + LSTM + TCN)
subprocess.run([
    'python', '-m', 'gridpulse.forecasting.train',
    '--config', 'configs/train_forecast.yaml'
], check=True)
```
